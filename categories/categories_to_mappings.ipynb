```python
import json
import re


def slugify(name):
    return re.sub(r"[^a-z0-9]+", "_", name.lower()).strip("_")


def build_full_category_dict(category_data):
    result = {}
    for main_category, pairs in category_data.items():
        if not pairs:
            continue

        display_name = None
        subcategories = {}

        for main_cat, sub_cat in pairs:
            if sub_cat.upper() == "UNKNOWN":
                continue
            if display_name is None:
                display_name = (
                    sub_cat  # first valid subcategory is used as display name
                )

            sub_key = slugify(sub_cat)
            subcategories[sub_key] = {
                "name": f"{main_cat}_{sub_key}",
                "displayName": sub_cat,
                "match_categories": [[main_cat, sub_cat]],
            }

        if display_name:
            result[main_category] = {
                "name": main_category,
                "displayName": display_name,
                "subcategories": subcategories,
            }

    return result


# Load input data
with open("category_summary.json", "r") as f:
    category_data = json.load(f)

# Build structured category mappings
new_category_mappings = build_full_category_dict(category_data)

# Save to JSON file
with open("new_category_mappings.json", "w") as f:
    json.dump(new_category_mappings, f, indent=2)
```
